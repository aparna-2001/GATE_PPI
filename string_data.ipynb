```python
import pandas as pd
```

```python
df = pd.read_csv(
    '/content/9606.protein.links.detailed.v12.0.onlyAB.tsv',
    sep='\t'
)

# Filter on experiments column, NOT combined_score
positives = df[df['experimental'] >= 700]
```

```python
import os
os.makedirs('data/raw', exist_ok=True)

# 1. TSV — human-readable, shareable, for inspection and reproducibility
#    You can open this in Excel, share with collaborators, submit with paper
positives.to_csv('data/raw/positives_string.tsv', sep='\t', index=False)

# 2. Parquet — compressed binary, fast to reload in Python for downstream steps
#    ~5x smaller than TSV, ~10x faster to load with pd.read_parquet()
positives.to_parquet('data/raw/positives_string.parquet', index=False)

# Verify both saved correctly
print(f"Rows saved: {len(positives):,}")
print(f"TSV size:     {os.path.getsize('data/raw/positives_string.tsv') / 1e6:.1f} MB")
print(f"Parquet size: {os.path.getsize('data/raw/positives_string.parquet') / 1e6:.1f} MB")
```

```python
import json
from datetime import datetime

metadata = {
    "source": "STRING v12.0",
    "file": "9606.protein.physical.links.detailed.v12.0.onlyAB.tsv.gz",
    "organism": "Homo sapiens (9606)",
    "filter_column": "experimental",
    "filter_threshold": 700,
    "date_downloaded": datetime.today().strftime('16-06-2026'),
    "rows_before_filter": len(df),
    "rows_after_filter": len(positives),
    "label": 1
}

with open('data/raw/positives_string_metadata.json', 'w') as f:
    json.dump(metadata, f, indent=2)

print(json.dumps(metadata, indent=2))
```
